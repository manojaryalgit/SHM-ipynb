# Simple Harmonic Motion (SHM) Simulator

This notebook simulates and visualizes Simple Harmonic Motion (SHM) of a mass-spring system. You'll see detailed plots of:
- Displacement, velocity, and acceleration over time
- Phase space diagram (displacement vs. velocity)
- Energy distribution (kinetic, potential, and total)

```python
import numpy as np
import matplotlib.pyplot as plt

# Plotting style
plt.style.use('ggplot')
%matplotlib inline
```

## Parameters and Initialization

```python
# Physical parameters
m = 0.2         # Mass (kg)
k = 1.0         # Spring constant (N/m)
A = 0.1         # Amplitude (m)
phi = 0         # Phase angle (rad)

# Derived quantities
omega = np.sqrt(k/m)  # Angular frequency (rad/s)
T = 2*np.pi/omega     # Period (s)

# Time array
t = np.linspace(0, 5*T, 1000)  # Simulate for 5 periods
```

## Equations of Motion

```python
x = A * np.cos(omega * t + phi)
v = -A * omega * np.sin(omega * t + phi)
a = -A * omega**2 * np.cos(omega * t + phi)
```

## Displacement, Velocity, and Acceleration vs. Time

```python
plt.figure(figsize=(12, 8))
plt.plot(t, x, label='Displacement x(t)')
plt.plot(t, v, label='Velocity v(t)')
plt.plot(t, a, label='Acceleration a(t)')
plt.xlabel('Time (s)')
plt.ylabel('Value')
plt.title('SHM Motion Parameters')
plt.legend()
plt.grid(True)
plt.show()
```

## Phase Space Plot (x vs. v)

```python
plt.figure(figsize=(8, 6))
plt.plot(x, v)
plt.xlabel('Displacement x (m)')
plt.ylabel('Velocity v (m/s)')
plt.title('Phase Space Diagram')
plt.grid(True)
plt.show()
```

## Energy Analysis

```python
KE = 0.5 * m * v**2
PE = 0.5 * k * x**2
TE = KE + PE

plt.figure(figsize=(12, 8))
plt.plot(t, KE, label='Kinetic Energy')
plt.plot(t, PE, label='Potential Energy')
plt.plot(t, TE, label='Total Energy', linestyle='--')
plt.xlabel('Time (s)')
plt.ylabel('Energy (J)')
plt.title('Energy vs Time in SHM')
plt.legend()
plt.grid(True)
plt.show()
```

## Conclusion

This simulation demonstrates key features of Simple Harmonic Motion:
- The displacement, velocity, and acceleration are sinusoidal.
- The phase space is an ellipse.
- Total energy is conserved and oscillates between kinetic and potential forms.
